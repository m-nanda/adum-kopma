```python
import pandas as pd
```

```python
# Load data dari excel
data = pd.read_excel('Data Sintetis Poin Keaktifan.xlsx')

# melihat 5 baris teratas data
data.head()
```

```python
# melihat 5 baris terbawah data
data.tail()
```

#### Terlihat bahwa daftar anggota hanya sampai baris ke-23. Maka, baris sisanya akan dibuang.

```python
# Membuang beberapa baris data terakhir yang tidak akan digunakan
data = data.iloc[:50,:]
```

```python
# Melihat keterangan kolom
data.info()
```

```python
# Mengambil anggota aktif saja
data = data[data['Aktif?']=='Ya']
data
```

```python
# Membuat Dataframe baru untuk hasil
data_hasil = data[['Nama', 'ID_Anggota', 'Jumlah Poin']]
data_hasil['Jumlah Poin'] = data_hasil['Jumlah Poin'].astype(int)
data_hasil.head()
```

```python
# Pilih N
N = 5

# Meyiapkan list / tempat penyimpanan sequence kegiatan anggota
list_kegiatan = []

# Looping untuk mengecek setiap anggota
for i in range(data.shape[0]):
    
    # Untuk nilai awal saja, agar bisa membuat batas maks. 3
    k = 0
    
    # Untuk menyimpan maks. 3 kegiatan anggota sebelum dimasukkan ke "list_kegiatan"
    kegiatan = []
    
    # Looping untuk mengecek kehadiran pada setiap kegiatan untuk setiap anggota 
    for j in range(3, data.shape[1]-1):
        
        # Untuk membatasi hanya 3 kegiatan, maka k < 3
        if data.iloc[i,j] == 1 and k < N:            
            kegiatan.append(data.columns[j]+'')
            k += 1
        
        # Untuk efisiensi kerja algoritma
        if k == N:
            break
    
    # Memasukkan kegiatan ke list_kegiatan
    list_kegiatan.append(', '.join(kegiatan))

data_hasil[str(N)+'-Kegiatan Terakhir yang Diikuti'] = list_kegiatan
data_hasil
```

```python
data_hasil.to_excel('Hasil ' +str(N) +'-Kegiatan Anggota.xlsx', index=False)
```
